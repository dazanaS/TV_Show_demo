## ğŸ“ˆ Visualizations: Netflix Originals Analysis


````python
# Visualization 1: IMDB Score Distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# IMDB Score Histogram
axes[0].hist(df_netflix_originals['IMDB Score'], bins=20, color='#E50914', edgecolor='white', alpha=0.8)
axes[0].set_xlabel('IMDB Score', fontsize=12)
axes[0].set_ylabel('Count', fontsize=12)
axes[0].set_title('Distribution of IMDB Scores', fontsize=14, fontweight='bold')
axes[0].axvline(df_netflix_originals['IMDB Score'].mean(), color='yellow', linestyle='--', 
                label=f'Mean: {df_netflix_originals["IMDB Score"].mean():.2f}')
axes[0].legend()

# Runtime Distribution
axes[1].hist(df_netflix_originals['Runtime'], bins=20, color='#564d4d', edgecolor='white', alpha=0.8)
axes[1].set_xlabel('Runtime (minutes)', fontsize=12)
axes[1].set_ylabel('Count', fontsize=12)
axes[1].set_title('Distribution of Runtime', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
````

````python
# Visualization 2: Top Genres
genre_counts = df_netflix_originals['Genre'].value_counts().head(15)

fig, ax = plt.subplots(figsize=(12, 6))
bars = ax.barh(genre_counts.index, genre_counts.values, color='#E50914', edgecolor='white')
ax.set_xlabel('Number of Titles', fontsize=12)
ax.set_title('Top 15 Genres in Netflix Originals', fontsize=14, fontweight='bold')
ax.invert_yaxis()

# Add value labels
for bar, val in zip(bars, genre_counts.values):
    ax.text(val + 1, bar.get_y() + bar.get_height()/2, str(val), va='center', fontsize=10)

plt.tight_layout()
plt.show()
````

````python
# Visualization 3: Releases Over Time
df_netflix_originals['Premiere'] = pd.to_datetime(df_netflix_originals['Premiere'], errors='coerce')
df_netflix_originals['Year'] = df_netflix_originals['Premiere'].dt.year

yearly_releases = df_netflix_originals['Year'].value_counts().sort_index()

fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(yearly_releases.index, yearly_releases.values, marker='o', linewidth=2, markersize=8, color='#E50914')
ax.fill_between(yearly_releases.index, yearly_releases.values, alpha=0.3, color='#E50914')
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Number of Releases', fontsize=12)
ax.set_title('Netflix Originals Releases by Year', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

## ğŸ’° Revenue Estimation: Stranger Things US Economic Impact

Since Netflix doesn't publish show-level revenue, here's how to estimate economic impact:


````python
# Revenue Estimation Example - Stranger Things
# Based on publicly available data from Netflix reports and industry sources

stranger_things_data = {
    'Season': [1, 2, 3, 4],
    'Release_Year': [2016, 2017, 2019, 2022],
    'Episodes': [8, 9, 8, 9],
    'Hours_Viewed_Billions': [0.5, 0.6, 0.8, 1.35],  # Estimated from Netflix reports
    'Production_Budget_Millions': [48, 72, 80, 270],  # Estimated from industry reports
    'US_Share_Percent': [40, 38, 35, 33],  # US share of viewership
}

df_st = pd.DataFrame(stranger_things_data)

# Calculate estimated US revenue contribution
# Assumption: 1 billion hours â‰ˆ $500M revenue contribution (rough estimate based on ARPU)
df_st['Est_US_Revenue_Millions'] = (
    df_st['Hours_Viewed_Billions'] * 
    (df_st['US_Share_Percent'] / 100) * 
    500  # $500M per billion hours
)

df_st['ROI_Percent'] = ((df_st['Est_US_Revenue_Millions'] - df_st['Production_Budget_Millions']) / 
                        df_st['Production_Budget_Millions'] * 100).round(1)

print("="*80)
print("ğŸ’° STRANGER THINGS: ESTIMATED US REVENUE BY SEASON")
print("="*80)
display(df_st)
````

````python
# Visualization: Revenue by Season
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Revenue Chart
colors = ['#E50914', '#B81D24', '#831010', '#5C0A0A']
bars1 = axes[0].bar(df_st['Season'].astype(str), df_st['Est_US_Revenue_Millions'], 
                    color=colors, edgecolor='white', linewidth=2)
axes[0].set_xlabel('Season', fontsize=12)
axes[0].set_ylabel('Estimated US Revenue ($M)', fontsize=12)
axes[0].set_title('Stranger Things: Est. US Revenue by Season', fontsize=14, fontweight='bold')

for bar, val in zip(bars1, df_st['Est_US_Revenue_Millions']):
    axes[0].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 5, 
                 f'${val:.0f}M', ha='center', fontsize=11, fontweight='bold')

# Budget vs Revenue
x = range(len(df_st))
width = 0.35
axes[1].bar([i - width/2 for i in x], df_st['Production_Budget_Millions'], 
            width, label='Budget', color='#564d4d', edgecolor='white')
axes[1].bar([i + width/2 for i in x], df_st['Est_US_Revenue_Millions'], 
            width, label='Est. US Revenue', color='#E50914', edgecolor='white')
axes[1].set_xlabel('Season', fontsize=12)
axes[1].set_ylabel('Amount ($M)', fontsize=12)
axes[1].set_title('Budget vs Estimated US Revenue', fontsize=14, fontweight='bold')
axes[1].set_xticks(x)
axes[1].set_xticklabels(['S1', 'S2', 'S3', 'S4'])
axes[1].legend()

plt.tight_layout()
plt.show()
````
